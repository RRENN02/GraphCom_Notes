```python
# Imports
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from mpl_toolkits.mplot3d.art3d import Poly3DCollection
```

1. Translate the prism in the last activity using the parameter [5, 7, 2]

```python
# Define the parameters for the rectangular prism
prism_base_length = 21
prism_base_width = 16
prism_height = 23

# Translation parameters
translation = np.array([5, 7, 2])

# Define the vertices of the rectangular prism
prism_vertices = np.array([
    [0, 0, 0],  # Base corner 1
    [prism_base_length, 0, 0],  # Base corner 2
    [prism_base_length, prism_base_width, 0],  # Base corner 3
    [0, prism_base_width, 0],  # Base corner 4
    [0, 0, prism_height],  # Top corner 1
    [prism_base_length, 0, prism_height],  # Top corner 2
    [prism_base_length, prism_base_width, prism_height],  # Top corner 3
    [0, prism_base_width, prism_height]  # Top corner 4
])

# Translate the vertices of the prism
translated_prism_vertices = prism_vertices + translation

# Define the faces of the rectangular prism using vertex indices
prism_faces = [
    [prism_vertices[0], prism_vertices[1], prism_vertices[2], prism_vertices[3]],  # Base
    [prism_vertices[4], prism_vertices[5], prism_vertices[6], prism_vertices[7]],  # Top
    [prism_vertices[0], prism_vertices[1], prism_vertices[5], prism_vertices[4]],  # Side 1
    [prism_vertices[1], prism_vertices[2], prism_vertices[6], prism_vertices[5]],  # Side 2
    [prism_vertices[2], prism_vertices[3], prism_vertices[7], prism_vertices[6]],  # Side 3
    [prism_vertices[3], prism_vertices[0], prism_vertices[4], prism_vertices[7]]   # Side 4
]

translated_prism_faces = [
    [translated_prism_vertices[0], translated_prism_vertices[1], translated_prism_vertices[2], translated_prism_vertices[3]],  # Base
    [translated_prism_vertices[4], translated_prism_vertices[5], translated_prism_vertices[6], translated_prism_vertices[7]],  # Top
    [translated_prism_vertices[0], translated_prism_vertices[1], translated_prism_vertices[5], translated_prism_vertices[4]],  # Side 1
    [translated_prism_vertices[1], translated_prism_vertices[2], translated_prism_vertices[6], translated_prism_vertices[5]],  # Side 2
    [translated_prism_vertices[2], translated_prism_vertices[3], translated_prism_vertices[7], translated_prism_vertices[6]],  # Side 3
    [translated_prism_vertices[3], translated_prism_vertices[0], translated_prism_vertices[4], translated_prism_vertices[7]]   # Side 4
]

# Create the 3D plot with two subplots
fig = plt.figure(figsize=(14, 6))

# Original prism subplot
ax1 = fig.add_subplot(121, projection='3d')
ax1.add_collection3d(Poly3DCollection(prism_faces, facecolors='orange', linewidths=1, edgecolors='red', alpha=.30))
ax1.set_title('Original Prism')
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_zlabel('Z')
ax1.set_xlim([0, prism_base_length + 10])
ax1.set_ylim([0, prism_base_width + 10])
ax1.set_zlim([0, prism_height + 10])

# Translated prism subplot
ax2 = fig.add_subplot(122, projection='3d')
ax2.add_collection3d(Poly3DCollection(translated_prism_faces, facecolors='orange', linewidths=1, edgecolors='red', alpha=.30))
ax2.set_title('Translated Prism')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_zlabel('Z')
ax2.set_xlim([0, prism_base_length + 10])
ax2.set_ylim([0, prism_base_width + 10])
ax2.set_zlim([0, prism_height + 10])

# Compute the volume of the rectangular prism
prism_volume = prism_base_length * prism_base_width * prism_height
print(f"Volume of the rectangular prism: {prism_volume}")

# Show the plot
plt.show()
```

2. Scale the cube in the last activity using the parameter [0, 2, 1]

```python
# Define the initial coordinate and edge length of the cube
initial_coord = (4, 8, 9)
edge_length = 12
scaling_factors = [0, 2, 1]

# Define the vertices of the cube based on the initial coordinate and edge length
vertices = np.array([
    [initial_coord[0], initial_coord[1], initial_coord[2]],  # Corner 1
    [initial_coord[0] + edge_length, initial_coord[1], initial_coord[2]],  # Corner 2
    [initial_coord[0] + edge_length, initial_coord[1] + edge_length, initial_coord[2]],  # Corner 3
    [initial_coord[0], initial_coord[1] + edge_length, initial_coord[2]],  # Corner 4
    [initial_coord[0], initial_coord[1], initial_coord[2] + edge_length],  # Corner 5
    [initial_coord[0] + edge_length, initial_coord[1], initial_coord[2] + edge_length],  # Corner 6
    [initial_coord[0] + edge_length, initial_coord[1] + edge_length, initial_coord[2] + edge_length],  # Corner 7
    [initial_coord[0], initial_coord[1] + edge_length, initial_coord[2] + edge_length]   # Corner 8
])

scaled_vertices = vertices * np.array(scaling_factors)

# Define the 6 faces of the cube using vertex indices
faces = [
    [vertices[0], vertices[1], vertices[2], vertices[3]],  # Bottom
    [vertices[4], vertices[5], vertices[6], vertices[7]],  # Top
    [vertices[0], vertices[1], vertices[5], vertices[4]],  # Front
    [vertices[2], vertices[3], vertices[7], vertices[6]],  # Back
    [vertices[1], vertices[2], vertices[6], vertices[5]],  # Right
    [vertices[4], vertices[7], vertices[3], vertices[0]]   # Left
]

scaled_faces = [
    [scaled_vertices[0], scaled_vertices[1], scaled_vertices[2], scaled_vertices[3]],  # Bottom
    [scaled_vertices[4], scaled_vertices[5], scaled_vertices[6], scaled_vertices[7]],  # Top
    [scaled_vertices[0], scaled_vertices[1], scaled_vertices[5], scaled_vertices[4]],  # Front
    [scaled_vertices[2], scaled_vertices[3], scaled_vertices[7], scaled_vertices[6]],  # Back
    [scaled_vertices[1], scaled_vertices[2], scaled_vertices[6], scaled_vertices[5]],  # Right
    [scaled_vertices[4], scaled_vertices[7], scaled_vertices[3], scaled_vertices[0]]   # Left
]

# Create the 3D plot with two subplots
fig = plt.figure(figsize=(14, 6))

# Original cube subplot
ax1 = fig.add_subplot(121, projection='3d')
ax1.add_collection3d(Poly3DCollection(faces, facecolors='blue', linewidths=1, edgecolors='red', alpha=.30))
ax1.set_title('Original Cube')
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_zlabel('Z')
ax1.set_xlim([vertices[:, 0].min() - 5, vertices[:, 0].max() + 5])
ax1.set_ylim([vertices[:, 1].min() - 5, vertices[:, 1].max() + 5])
ax1.set_zlim([vertices[:, 2].min() - 5, vertices[:, 2].max() + 5])

# Scaled cube subplot
ax2 = fig.add_subplot(122, projection='3d')
ax2.add_collection3d(Poly3DCollection(scaled_faces, facecolors='blue', linewidths=1, edgecolors='red', alpha=.30))
ax2.set_title('Scaled Cube')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_zlabel('Z')
ax2.set_xlim([scaled_vertices[:, 0].min() - 5, scaled_vertices[:, 0].max() + 5])
ax2.set_ylim([scaled_vertices[:, 1].min() - 5, scaled_vertices[:, 1].max() + 5])
ax2.set_zlim([scaled_vertices[:, 2].min() - 5, scaled_vertices[:, 2].max() + 5])

# Compute the volume of the original and scaled cubes
cube_volume = edge_length ** 3
scaled_volume = edge_length * scaling_factors[0] * edge_length * scaling_factors[1] * edge_length * scaling_factors[2]

print(f"Volume of the original cube: {cube_volume}")
print(f"Volume of the scaled cube: {scaled_volume}")  # results in 0 because the x scaling factor is 0

# Show the plot
plt.show()
```

3. Rotate the pyramid in the last activity using 120 degrees about x, y and z axis.

3.1  X axis
|   1   0       0       0   |
|   0   cos(0)  -sin(0) 0   |
|   0   sin(0)  cos(0)  0   |
|   0   0       0       1   |

```python
# Define the parameters for the pyramid
pyramid_base_length = 21
pyramid_base_width = 16
pyramid_height = 23
rotation_angle = (2*np.pi)/3  # Rotation angle in radians (120 degrees)

# Apply rotation around the x-axis
rotation_matrix = np.array([
    [1, 0, 0],
    [0, np.cos(rotation_angle), -np.sin(rotation_angle)],
    [0, np.sin(rotation_angle), np.cos(rotation_angle)]
])

# Define the vertices of the pyramid
pyramid_vertices = np.array([
    [0, 0, 0],  # Base corner 1
    [pyramid_base_length, 0, 0],  # Base corner 2
    [pyramid_base_length, pyramid_base_width, 0],  # Base corner 3
    [0, pyramid_base_width, 0],  # Base corner 4
    [pyramid_base_length / 2, pyramid_base_width / 2, pyramid_height]  # Apex
])

rotated_pyramid = np.dot(pyramid_vertices, rotation_matrix.T)  # Use .T for transpose which corrects the alignment of dimensions of the matrix

# Define the faces of the pyramid using vertex indices
pyramid_faces = [
    [pyramid_vertices[0], pyramid_vertices[1], pyramid_vertices[4]],  # Side 1
    [pyramid_vertices[1], pyramid_vertices[2], pyramid_vertices[4]],  # Side 2
    [pyramid_vertices[2], pyramid_vertices[3], pyramid_vertices[4]],  # Side 3
    [pyramid_vertices[3], pyramid_vertices[0], pyramid_vertices[4]],  # Side 4
    [pyramid_vertices[0], pyramid_vertices[1], pyramid_vertices[2], pyramid_vertices[3]]  # Base
]

rotated_pyramid_faces = [
    [rotated_pyramid[0], rotated_pyramid[1], rotated_pyramid[4]],  # Side 1
    [rotated_pyramid[1], rotated_pyramid[2], rotated_pyramid[4]],  # Side 2
    [rotated_pyramid[2], rotated_pyramid[3], rotated_pyramid[4]],  # Side 3
    [rotated_pyramid[3], rotated_pyramid[0], rotated_pyramid[4]],  # Side 4
    [rotated_pyramid[0], rotated_pyramid[1], rotated_pyramid[2], rotated_pyramid[3]]  # Base
]

# Create the 3D plot with two subplots
fig = plt.figure(figsize=(14, 6))

# Original pyramid subplot
ax1 = fig.add_subplot(121, projection='3d')
ax1.add_collection3d(Poly3DCollection(pyramid_faces, facecolors='green', linewidths=1, edgecolors='red', alpha=.30))
ax1.set_title('Original Pyramid')
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_zlabel('Z')
ax1.set_xlim([-20, 40])
ax1.set_ylim([-20, 40])
ax1.set_zlim([-20, 40])

# Rotated pyramid subplot
ax2 = fig.add_subplot(122, projection='3d')
ax2.add_collection3d(Poly3DCollection(rotated_pyramid_faces, facecolors='green', linewidths=1, edgecolors='red', alpha=.30))
ax2.set_title('Rotated Pyramid')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_zlabel('Z')
ax2.set_xlim([-20, 40])
ax2.set_ylim([-20, 40])
ax2.set_zlim([-20, 40])

# Compute the volume of the pyramid
pyramid_volume = (1/3) * pyramid_base_length * pyramid_base_width * pyramid_height
print(f"Volume of the pyramid: {pyramid_volume}")

# Show the plot
plt.show()
```

Combine all rotations to a single pyramid

```python
# Define the parameters for the pyramid
pyramid_base_length = 21
pyramid_base_width = 16
pyramid_height = 23
rotation_angle = (2*np.pi)/3  # Rotation angle in radians (120 degrees)

# Define the rotation matrices for x, y, and z axes
rotation_matrix_x = np.array([
    [1, 0, 0],
    [0, np.cos(rotation_angle), -np.sin(rotation_angle)],
    [0, np.sin(rotation_angle), np.cos(rotation_angle)]
])

rotation_matrix_y = np.array([
    [np.cos(rotation_angle), 0, np.sin(rotation_angle)],
    [0, 1, 0],
    [-np.sin(rotation_angle), 0, np.cos(rotation_angle)]
])

rotation_matrix_z = np.array([
    [np.cos(rotation_angle), -np.sin(rotation_angle), 0],
    [np.sin(rotation_angle), np.cos(rotation_angle), 0],
    [0, 0, 1]
])

# Define the vertices of the pyramid
pyramid_vertices = np.array([
    [0, 0, 0],  # Base corner 1
    [pyramid_base_length, 0, 0],  # Base corner 2
    [pyramid_base_length, pyramid_base_width, 0],  # Base corner 3
    [0, pyramid_base_width, 0],  # Base corner 4
    [pyramid_base_length / 2, pyramid_base_width / 2, pyramid_height]  # Apex
])

# Apply the rotations sequentially
rotated_pyramid = pyramid_vertices @ rotation_matrix_x.T
rotated_pyramid = rotated_pyramid @ rotation_matrix_y.T
rotated_pyramid = rotated_pyramid @ rotation_matrix_z.T

# Define the faces of the pyramid using vertex indices
pyramid_faces = [
    [pyramid_vertices[0], pyramid_vertices[1], pyramid_vertices[4]],  # Side 1
    [pyramid_vertices[1], pyramid_vertices[2], pyramid_vertices[4]],  # Side 2
    [pyramid_vertices[2], pyramid_vertices[3], pyramid_vertices[4]],  # Side 3
    [pyramid_vertices[3], pyramid_vertices[0], pyramid_vertices[4]],  # Side 4
    [pyramid_vertices[0], pyramid_vertices[1], pyramid_vertices[2], pyramid_vertices[3]]  # Base
]

rotated_pyramid_faces = [
    [rotated_pyramid[0], rotated_pyramid[1], rotated_pyramid[4]],  # Side 1
    [rotated_pyramid[1], rotated_pyramid[2], rotated_pyramid[4]],  # Side 2
    [rotated_pyramid[2], rotated_pyramid[3], rotated_pyramid[4]],  # Side 3
    [rotated_pyramid[3], rotated_pyramid[0], rotated_pyramid[4]],  # Side 4
    [rotated_pyramid[0], rotated_pyramid[1], rotated_pyramid[2], rotated_pyramid[3]]  # Base
]

# Create the 3D plot with two subplots
fig = plt.figure(figsize=(14, 6))

# Original pyramid subplot
ax1 = fig.add_subplot(121, projection='3d')
ax1.add_collection3d(Poly3DCollection(pyramid_faces, facecolors='green', linewidths=1, edgecolors='red', alpha=.30))
ax1.set_title('Original Pyramid')
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_zlabel('Z')
ax1.set_xlim([-20, 40])
ax1.set_ylim([-20, 40])
ax1.set_zlim([-20, 40])

# Rotated pyramid subplot
ax2 = fig.add_subplot(122, projection='3d')
ax2.add_collection3d(Poly3DCollection(rotated_pyramid_faces, facecolors='green', linewidths=1, edgecolors='red', alpha=.30))
ax2.set_title('Rotated Pyramid (Sequential Axes)')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_zlabel('Z')
ax2.set_xlim([-20, 40])
ax2.set_ylim([-20, 40])
ax2.set_zlim([-20, 40])

# Compute the volume of the pyramid
pyramid_volume = (1/3) * pyramid_base_length * pyramid_base_width * pyramid_height
print(f"Volume of the pyramid: {pyramid_volume}")

# Show the plot
plt.show()
```
