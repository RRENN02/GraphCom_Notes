```python
# Imports
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from mpl_toolkits.mplot3d.art3d import Poly3DCollection
```

1. Create a 3D line plot that shows the path of a sine wave in 3D space. 

```python
# Create a new figure for the 3D plot with a specified size
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# sine wave
z = np.linspace(0, 2, 100)
x = np.sin(2 * np.pi * z)
y = np.cos(2 * np.pi * z)

# Plot data
ax.plot(x, y, z, label='Sine Wave')

# Add legend
ax.legend()

# Labels
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.show()
```

Sine Wave Ripple Effect

```python
# Create a new figure for the 3D plot with a specified size
# Create a grid in the x-y plane
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)

# Calculate the sine wave ripple effect
z = np.sin(np.sqrt(x**2 + y**2))

# Create a new figure for the 3D plot with a specified size
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
ax.plot_surface(x, y, z, cmap='viridis')

# Labels
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

ax = fig.suptitle('Sine Wave Ripple Effect')

plt.show()
```
